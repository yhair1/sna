## Network interaction

```python
attributes = pd.read_excel ('NIC_Clean Data.xlsx', sheet_name='interaction')
```

```python
attributes = attributes[['Unnamed: 0', 'AASCU', 'ATD', 'AIHEC', 'APLU', 'CCA', 'E3', 'EXA', 'GI',
       'GIA', 'JFF', 'MDRC', 'NWCCU', 'UNCF',]]
```

```python
attributes = attributes.rename(columns={'Unnamed: 0':'index'}).set_index('index')
```

```python
attributes = attributes[:13]
```

```python
attributes
```

```python
node_names = attributes.columns
```

```python
M = attributes.values
#M = m.to_numpy()

# create graph
g = Graph.Adjacency(((M>0).tolist()), mode='directed', weighted=True)

# add edge weights and node labels.
g.es['weight'] = M[M.nonzero()]

# define network measures
network_measures = pd.DataFrame(dict(
    org = node_names,
    weight_in = g.indegree(),
    weight_out = g.outdegree(),
    in_degree = g.degree(mode='in'),
    out_degree = g.degree(mode='out'),
     betweenness = g.betweenness(weights='weight'),
    edge_count = g.ecount(),
    density = g.density() 

))
```

```python
#  define vertix size
network_measures['weight_in'] = network_measures['weight_in']

g.vs['size'] = network_measures['weight_out']*(4)
vertix_size = g.vs['size']
```

```python
# set vertix color
color = ["#88203A", "#008E7F", "#0091BA", "#046C4C",
                                  "#7E6EFF", "#268E3F", "#D73246", "#243788",
                                  "#D06338", "#FF532E", "#085792", "#264F34",
                                  "#F49722"] 
g.vs['color'] = color

vertix_color = g.vs['color']

#set edge color that match with vertix color

TEMP=[]
for e in g.es:
    elist = list(e.tuple)
    TEMP.append(elist)
g.es["color_assigned"] = TEMP
color_dict = {"0":"#88203A", "1":"#008E7F", "2":"#0091BA", "3":"#046C4C",
                                  "4":"#7E6EFF", "5":"#268E3F", "6":"#D73246", "7":"#243788",
                                  "8":"#D06338", "9":"#FF532E", "10":"#085792", "11":"#264F34", 
                                  "12":"#F49722"}
color_assigned = []
for i in range(129):
    color_assigned.append(g.es["color_assigned"][i][0])

edge_color = [color_dict[str(int(col_num))] for col_num in color_assigned]


# set vertix label
g.vs['label'] = node_names


## set layout
g_cr = g.layout_circle()
layout = g_cr
#set visual styel 
visual_style = {}
out_name = "interaction_cr_new.png"
visual_style["vertex_color"] = g.vs['color']
visual_style["vertex_size"] = g.vs['size']
visual_style["vertex_frame_color"] = g.vs['color']
visual_style["vertex_label_color"] = 'white'
visual_style["vertex_label_cex"] = 0.7
visual_style["vertex_label_family"] = 'sans'
visual_style["vertex_label_font"] = 2
visual_style["edge_color"] = edge_color
visual_style["layout"] = layout
visual_style["edge_lty"] = 1
visual_style["edge_arrow_size"] = 0.6
visual_style["edge_width"] = g.es['weight']
visual_style["edge_curved"] = 0.2
visual_style["bbox"] = (400, 400)
visual_style["margin"] = 40
# export network measures in csv
network_measures.to_csv('network_measures_new.csv')
# plot the graph showing the entire network interaction
plot(g, out_name, **visual_style)
```
